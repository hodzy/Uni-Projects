```python
# here we are using contouring to find the angle of the image, and then using it to make the image upright

def random_rotate_base_test(path, angle_range):
    
    # dimenstions for resizing (width, height)
    dim = (28, 28)

    data = []
    labels = []
    valid_images = [".jpg"]

    for i in range(10):
        path = path + "\\" + str(i)
#     print(path)
        print(i)
        for f in os.listdir(path):
            ext = os.path.splitext(f)[1]
            if ext.lower() not in valid_images:
                continue
            image = cv2.imread(os.path.join(path,f))
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            
            random_angle = random.randint(-angle_range, angle_range) 
#             print(random_angle)
            rotated_image = imutils.rotate_bound(gray, angle= random_angle)
            
            

            
            # test
            ret,thresh = cv2.threshold(rotated_image,50,255,0)
            resized = cv2.resize(thresh, dim, interpolation = cv2.INTER_AREA)
            test = resized.flatten()            
            

            
            
            data.append(test)
            labels.append(i)
     
        path = path.rsplit('\\',1)[0]

    return (data, labels)
```

```python
print("preprocessing: random rotate training data")
taining_path = ".\MNIST_Dataset_JPG_format\MNIST_JPG_training"
(training_data, training_labels) = read_train(taining_path)
```

```python
from sklearn.ensemble import RandomForestClassifier



# change max depth
# clf = RandomForestClassifier(max_depth=15, random_state=0)
# clf = RandomForestClassifier(max_depth=30, random_state=0)
clf = RandomForestClassifier(max_depth=50, random_state=0)


clf.fit(training_data, training_labels)
```

```python
random_angle_test = 0
print("preprocessing: random rotate test data")
test_path = ".\MNIST_Dataset_JPG_format\MNIST_JPG_testing"
(test_data_base, test_labels_base) = random_rotate_base_test(test_path, random_angle_test)
```

```python
test_results_base = clf.predict(test_data_base)
```

```python
print(test_results_base)
```

```python
error = 0
for i in range(len(test_results_base)):
    if test_results_base[i] != test_labels_base[i]:
        error += 1
print("errors:", error)
print("sample length: ", len(test_labels_base))
print("accuracy: ", (len(test_labels_base)-error)/len(test_labels_base)*100)
```

```python
def calculate_accuracy(test_results_base, test_labels_base):
    error = 0
    for i in range(len(test_results_base)):
        if test_results_base[i] != test_labels_base[i]:
            error += 1
    return ((len(test_labels_base)-error)/len(test_labels_base)*100)
```

```python
# added later
print("preprocessing: random rotate test data")
test_path = ".\MNIST_Dataset_JPG_format\MNIST_JPG_testing"
results = []

for angle in range(0, 136, 15): 
    print("angle: ", angle)
    (test_data_base, test_labels_base) = random_rotate_base_test(test_path, angle)
    test_results_base = clf.predict(test_data_base)
    results.append(calculate_accuracy(test_results_base, test_labels_base))
```
